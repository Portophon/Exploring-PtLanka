## Packages

```python
import pandas as pd
```

# Data sets

```python
wikcionario_IPA = pd.read_csv("wikcionario_IPA_final.csv")
wikcionario_IPA.columns = ["Portuguese_definition","Portuguese_IPA"]
wikcionario_IPA
```

```python
PtLanka = pd.read_csv("PtLanka.csv")
PtLanka
```

# Merge

```python
PtLanka = pd.merge(PtLanka, wikcionario_IPA, on="Portuguese_definition", how='left')
PtLanka
```

```python
col = PtLanka.pop("Portuguese_IPA")
PtLanka.insert(3, col.name, col)
PtLanka
```

```python
PtLanka.to_csv('PtLanka-final.csv', na_rep='')
```
